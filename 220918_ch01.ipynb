### 1.3.3 특성과 타깃 준비하기

```python
# x는 input, y는 output
X = df_bikes.iloc[:,:-1]
y = df_bikes.iloc[:,-1]
```

### 1.3.4 회귀 이해하기

### 1.3.5 사이킷런 사용하기

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# train, test data 분리
# random_state=2 -> 의사 난수 생성기 를 위한 시드값 지정!
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

### 1.3.6 경고 끄기

```python
# 사이킷런은 버전 변경사항에 대해 경고를 보내는데, 우선은 무시하자
import warnings
warnings.filterwarnings('ignore')
```

### 1.3.7 선형 회귀 모델 만들기

```python
# 머신러닝 모델 객체 생성
lin_reg = LinearRegression()
```

```python
# 모델을 훈련세트에서 훈련
lin_reg.fit(X_train, y_train)
```

```python
# 테스트 세트에서 예측 출력
y_pred = lin_reg.predict(X_test)
```

```python
# 실제값과 대사
# 선형회귀에서는 평균 제곱근 오차 (RMSE) 사용 (MSE필요)

from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
# 결과 출력
print("RMSE : %0.2f" % (rmse))
```

### 1.3.10 교차검증

선형 회귀 교차 검증

```python
from sklearn.model_selection import cross_val_score
```

```python
model = LinearRegression()
```

```python
# 폴드개수 10개로 적용
# 왜 neg_~ 이 이유 : 사이킷런은 점수가 높을수록 더 좋은 모델이라고 판단함! -> mse와는 반대이기떄문에 neg 붙임
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)
```

```python
rmse = np.sqrt(-scores)
```

```python
print('회귀 rmse:', np.round(rmse,2))
print('RMSE 평균: %0.2f' % (rmse.mean()))
```

XGBoost 교차 검증

```python
model = XGBRegressor()
```

```python
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)
```

```python
rmse = np.sqrt(-scores)
```

```python
print('회귀 rmse:', np.round(rmse,2))
print('RMSE 평균: %0.2f' % (rmse.mean()))
```

### 1.4.3 데이터 랭글링

```python
# 데이터 적재
# 칼럼명이 없기 때문에 headr는 none(첫번째 행부터 바로 데이터)
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
```

```python
df_census.head()
```

```python
df_census.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                  'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 
                   'income']
df_census.head()
```

```python
# 누락된 값 확인
df_census.info()
```

```python
# education 열 삭제
df_census = df_census.drop(['education'], axis = 1)
```

```python
# 수치형이 아닌 열을 수치형으로 변환 -> pd.get_dummies() 함수 활요!
df_census = pd.get_dummies(df_census)
df_census.head()
```

```python
df_census = df_census.drop('income_ <=50K', axis = 1)
```

```python
# 특성데이터, 타겟데이터 분할
X = df_census.iloc[:,:-1]
y = df_census.iloc[:,-1]
```

## 1.4.4 로지스틱 회귀

```python
# 로지스틱 회귀는 각 특성에 대해 모델 정확도를 최대화하는 적절한 가중치 또는 계수를 찾음
# 사이킷런으로 로지스틱 회귀 모델을 만드는 것은 선형회귀와 거의 동이
# 사이킷런에서 분류모델의 경우 기본적으로 정확도로 모델을 평가하므로 scoring 매개변수를 지정할 필요 없음

from sklearn.linear_model import LogisticRegression
```

```python
# 정확도 출력하는 교차검증 함수 생성

def cross_val(classifier, num_splits=10):
  model = classifier
  scores = cross_val_score(model, X, y, cv=num_splits)
  print('정확도:', np.round(scores, 2))
  print('평균 정확도: %0.2f' % (scores.mean()))
                           
```

```python
cross_val(LogisticRegression())
```

### 1.4.5 XGBClassifier

```python
from xgboost import XGBClassifier
```

```python
cross_val(XGBClassifier(n_estimators=5))
```
